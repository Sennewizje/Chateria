% Reguła sprawdzająca, czy istnieje bezpośrednie połączenie lotnicze między dwoma lotniskami

    polaczenie(X, Y) :- polaczenie(X, Y, _).


% Reguła określająca koszt podróży między dwoma lotniskami

    koszt_podrozy(X, Y, C) :- polaczenie(X, Y, C).

% Znajdowanie najkrótszej trasy za pomocą przeszukiwania wszerz, wraz z liczbą przesiadek

    najkrotsza_trasa(Start, Cel, Trasa, LiczbaPrzesiadek) :-
        przeszukiwanie_wszerz([[Start]], Cel, TrasaReversed),
        reverse(TrasaReversed, Trasa),
        length(Trasa, DlugoscTrasy),
        LiczbaPrzesiadek is DlugoscTrasy - 1.

    przeszukiwanie_wszerz([[Cel|Odwiedzone]|_], Cel, [Cel|Odwiedzone]).

    przeszukiwanie_wszerz([[Bieżący|Odwiedzone]|Kolejka], Cel, Trasa) :-
        findall([Sąsiad, Bieżący|Odwiedzone],
                (polaczenie(Bieżący, Sąsiad, _), \+ member(Sąsiad, [Bieżący|Odwiedzone])),
                NoweSciezki),
        append(Kolejka, NoweSciezki, ZaktualizowanaKolejka),
        przeszukiwanie_wszerz(ZaktualizowanaKolejka, Cel, Trasa).



% Znajdowanie najtańszej drogi z użyciem strategii zachłannej

    najtansza_droga(Start, Cel, Trasa, Koszt) :-
        znajdz_najtansza([(0, [Start])], Cel, Trasa, Koszt).

    znajdz_najtansza([(Koszt, [Cel|OdwroconaTrasa])|_], Cel, Trasa, Koszt) :-
        reverse([Cel|OdwroconaTrasa], Trasa).

    znajdz_najtansza([(ObecnyKoszt, [Bieżący|Odwiedzone])|ResztaKolejki], Cel, OptymalnaTrasa, MinimalnyKoszt) :-
        findall(
            (NowyKoszt, [Sąsiad|[Bieżący|Odwiedzone]]),
            (polaczenie(Bieżący, Sąsiad, Dystans), \+ member(Sąsiad, Odwiedzone), NowyKoszt is ObecnyKoszt + Dystans),
            MozliweDrogi
        ),
        append(ResztaKolejki, MozliweDrogi, KolejkaZaktualizowana),
        sort(KolejkaZaktualizowana, KolejkaPosortowana), % Sortowanie według kosztu
        znajdz_najtansza(KolejkaPosortowana, Cel, OptymalnaTrasa, MinimalnyKoszt).

% Sprawdza, czy produkt występuje w ilości > 0

      dostepny_produkt(Nazwa, Produkty) :-
          member(produkt(Nazwa, Ilosc, _), Produkty),
          Ilosc > 0.

% sprawdzanie czy dany produkt jest w magazynie

      czy_dostepny(Nazwa) :-
          produkty(Produkty),
          dostepny_produkt(Nazwa, Produkty).

% Dodawanie produktu do listy

      dodaj_produkt(Produkt, StaraLista, NowaLista) :-
          append([Produkt], StaraLista, NowaLista).

% Usuwanie produktu z listy

      usun_produkt(Produkt, StaraLista, NowaLista) :-
          delete(StaraLista, Produkt, NowaLista).

% Reguła uzupełniająca ilość danego produktu

      uzupelnij_stan(Nazwa, Dodatek, Produkty, NoweProdukty) :-
          select(produkt(Nazwa, Ilosc, Sektor), Produkty, TymczasowaLista), % Znajdź i usuń stary produkt
          NowaIlosc is Ilosc + Dodatek, % Oblicz nową ilość
          append([produkt(Nazwa, NowaIlosc, Sektor)], TymczasowaLista, NoweProdukty).

% reguła uzupełniająca stan produktu i wypisyująca aktualny stan

      uzupelnij_i_wypisz(Nazwa, Dodatek) :-
          produkty(Lista), % Pobierz oryginalną listę produktów
          uzupelnij_stan(Nazwa, Dodatek, Lista, NowaLista), % Uzupełnij stan
          wypisz_produkty(NowaLista). % Wyświetl zaktualizowaną listę

%redula wyświetlająca produkty

      wypisz_produkty([]).
      wypisz_produkty([produkt(Nazwa, Ilosc, Sektor) | Reszta]) :-
          format('Produkt: ~w, Ilość: ~w, Sektor: ~w~n', [Nazwa, Ilosc, Sektor]),
          wypisz_produkty(Reszta).

% Usuwa wiele produktów z listy

      usun_wiele_produktow([], Lista, Lista). % Jeśli lista produktów do usunięcia jest pusta, zwróć listę bez zmian.
      usun_wiele_produktow([Produkt | Reszta], StaraLista, NowaLista) :-
          usun_produkt(Produkt, StaraLista, TymczasowaLista), % Usuń pierwszy produkt.
          usun_wiele_produktow(Reszta, TymczasowaLista, NowaLista). % Rekurencyjnie usuń pozostałe produkty.

% usuwanie wielu produktów na raz i wypisywanie aktualnego stanu

      usun_i_wypisz(ProduktyDoUsuniecia) :-
          produkty(Lista),
          usun_wiele_produktow(ProduktyDoUsuniecia, Lista, NowaLista),
          wypisz_produkty(NowaLista).

% Przenoszenie produktu do nowego sektora

      przenies_produkt(Nazwa, NowySektor, Produkty, NoweProdukty) :-
          select(produkt(Nazwa, Ilosc, _), Produkty, TymczasowaLista), % Znajdź i usuń produkt o nazwie Nazwa
          append([produkt(Nazwa, Ilosc, NowySektor)], TymczasowaLista, NoweProdukty). % Dodaj go z nowym sektorem

% Przeniesienie produktu do nowego sektora i wypisywanie aktualnego stanu

      przenies_i_wypisz(Nazwa, NowySektor) :-
          produkty(Lista), % Pobierz listę produktów
          przenies_produkt(Nazwa, NowySektor, Lista, NowaLista), % Przenieś produkt
          wypisz_produkty(NowaLista). % Wyświetl zaktualizowaną listę

**%Reguły**

% Reguła do przeliczania kilometrów na czas przejazdu (w godzinach)

    czas_z_kilometrow(Kilometry, Czas) :-
        SredniaPredkosc is 60, % Średnia prędkość w km/h
        Czas is Kilometry / SredniaPredkosc.

% Reguła do obliczania czasu przejazdu między dwoma miastami

    czas_podrozy(X, Y, Czas) :-
        droga(X, Y, Kilometry),
        czas_z_kilometrow(Kilometry, Czas).

    czas_podrozy(X, Y, Czas) :-
        droga(Y, X, Kilometry),
        czas_z_kilometrow(Kilometry, Czas).

% Pomocnicza reguła do obliczania ścieżek i czasu

    sciezka(X, Y, Czas, Trasa) :-
        sciezka_helper(X, Y, Czas, Trasa, [X]).

    sciezka_helper(X, Y, Czas, [X, Y], _) :-
        droga(X, Y, Kilometry),
        czas_z_kilometrow(Kilometry, Czas).

    sciezka_helper(X, Y, Czas, [X|ResztaTrasy], Odwiedzone) :-
        droga(X, Z, Kilometry),
        \+ member(Z, Odwiedzone),
        czas_z_kilometrow(Kilometry, CzasDoZ),
        sciezka_helper(Z, Y, CzasZDoY, ResztaTrasy, [Z|Odwiedzone]),
        Czas is CzasDoZ + CzasZDoY.

% Znajdowanie najkrótszej ścieżki pod względem czasu przejazdu

    najkrotsza_sciezka(X, Y, NajkrotszaTrasa, NajkrotszyCzas) :-
        findall((Czas, Trasa), sciezka(X, Y, Czas, Trasa), WszystkieTrasy),
        sort(WszystkieTrasy, [(NajkrotszyCzas, NajkrotszaTrasa)|_]).

% Wyświetlenie najkrótszej trasy

    pokaz_najkrotsza_sciezka(X, Y) :-
        najkrotsza_sciezka(X, Y, NajkrotszaTrasa, NajkrotszyCzas),
        format('Najkrótsza trasa: ~w, czas: ~2f godzin~n', [NajkrotszaTrasa, NajkrotszyCzas]).
